```python
# Kasus 1 – Hitung Tekanan (P)
# Logika Percabangan dan Fungsi
# Konversi otomatis satuan ke SI (Pa, m³, K)

# Konstanta gas ideal (dalam satuan SI)
R = 8.314  # J/(mol·K)

def konversi_tekanan(P, satuan):
    """Konversi tekanan ke Pascal (Pa)"""
    if satuan.lower() == "atm":
        return P * 101325
    elif satuan.lower() == "pa":
        return P
    else:
        print("Satuan tekanan tidak dikenali.")
        return None

def konversi_volume(V, satuan):
    """Konversi volume ke meter kubik (m³)"""
    if satuan.lower() == "l":
        return V / 1000
    elif satuan.lower() == "m3" or satuan.lower() == "m^3":
        return V
    else:
        print("Satuan volume tidak dikenali.")
        return None

def konversi_suhu(T, satuan):
    """Konversi suhu ke Kelvin (K)"""
    if satuan.lower() == "c" or satuan.lower() == "°c":
        return T + 273.15
    elif satuan.lower() == "k":
        return T
    else:
        print("Satuan suhu tidak dikenali.")
        return None

def main():
    print("=== PROGRAM PERSAMAAN GAS IDEAL ===")
    print("Persamaan: PV = nRT")
    print("Pilih variabel yang ingin dihitung:")
    print("1. Tekanan (P)")
    print("2. Volume (V)")
    print("3. Jumlah mol (n)")
    print("4. Suhu (T)")

    pilihan = input("Masukkan pilihan (1/2/3/4): ")

    try:
        # Hitung Tekanan (P)
        if pilihan == "1":
            V = float(input("Masukkan volume: "))
            satuan_V = input("Satuan volume (L/m3): ")
            V = konversi_volume(V, satuan_V)

            n = float(input("Masukkan jumlah mol (mol): "))

            T = float(input("Masukkan suhu: "))
            satuan_T = input("Satuan suhu (K/C): ")
            T = konversi_suhu(T, satuan_T)

            P = (n * R * T) / V
            print(f"\nTekanan (P) = {P:.2f} Pa")

        # Hitung Volume (V)
        elif pilihan == "2":
            P = float(input("Masukkan tekanan: "))
            satuan_P = input("Satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(P, satuan_P)

            n = float(input("Masukkan jumlah mol (mol): "))

            T = float(input("Masukkan suhu: "))
            satuan_T = input("Satuan suhu (K/C): ")
            T = konversi_suhu(T, satuan_T)

            V = (n * R * T) / P
            print(f"\nVolume (V) = {V:.6f} m³")

        # Hitung Jumlah mol (n)
        elif pilihan == "3":
            P = float(input("Masukkan tekanan: "))
            satuan_P = input("Satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(P, satuan_P)

            V = float(input("Masukkan volume: "))
            satuan_V = input("Satuan volume (L/m3): ")
            V = konversi_volume(V, satuan_V)

            T = float(input("Masukkan suhu: "))
            satuan_T = input("Satuan suhu (K/C): ")
            T = konversi_suhu(T, satuan_T)

            n = (P * V) / (R * T)
            print(f"\nJumlah mol (n) = {n:.6f} mol")

        # Hitung Suhu (T)
        elif pilihan == "4":
            P = float(input("Masukkan tekanan: "))
            satuan_P = input("Satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(P, satuan_P)

            V = float(input("Masukkan volume: "))
            satuan_V = input("Satuan volume (L/m3): ")
            V = konversi_volume(V, satuan_V)

            n = float(input("Masukkan jumlah mol (mol): "))

            T = (P * V) / (n * R)
            print(f"\nSuhu (T) = {T:.2f} K")

        else:
            print("Pilihan tidak valid.")

    except ZeroDivisionError:
        print("Error: Pembagian dengan nol tidak diperbolehkan.")
    except ValueError:
        print("Error: Input harus berupa angka.")

if __name__ == "__main__":
    main()
```
