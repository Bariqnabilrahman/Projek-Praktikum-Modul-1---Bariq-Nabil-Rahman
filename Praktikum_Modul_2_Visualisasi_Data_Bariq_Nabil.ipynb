```python
#Kasus 1 - Standar Jatuh Bebas
import matplotlib.pyplot as plt

h = 100
g = 9.8
t = list(range(11))
s = [h - 0.5 * g*ti**2 for ti in t]
v = [g*ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Gerak Jatuh Bebas (Kasus 1)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 1:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")
```

```python
import matplotlib.pyplot as plt
import numpy as np

v0 = 10
h = 120
g = 9.8
t = np.arange(0.5, 11.0, 0.5)
s = [h - 0.5 * g*ti**2 for ti in t]
v = [v0 + g*ti for ti in t]
a = [g] * len(t)

plt.plot(t, s, marker='o', label='Posisi (m)')
plt.plot(t, v, marker='s', label='Kecepatan (m/s)')
plt.plot(t, a, marker='^', label='Percepatan (m/s²)')
plt.title("Jatuh Bebas dengan Kecepatan Awal (Kasus 2)")
plt.xlabel("Waktu (detik)")
plt.ylabel("Nilai")
plt.grid(True)
plt.legend()
plt.show()

print("Kasus 2:")
print(f"Posisi maksimum: {max(s):.2f} m")
print(f"Kecepatan maksimum: {max(v):.2f} m/s")
```
