```python

def simulasi_gerak(m, a, v0, h0, t_total, dt):
    g = 9.8  # percepatan gravitasi (m/s^2)
    print(f"\n{'Waktu (s)':<10}{'Posisi (m)':<15}{'Kecepatan (m/s)':<20}{'Energi Kinetik (J)':<20}{'Energi Potensial (J)':<20}")
    print("-" * 85)

    t = 0
    while t <= t_total:
        s = h0 + v0 * t + 0.5 * a * t**2       # posisi
        v = v0 + a * t                         # kecepatan
        Ek = 0.5 * m * v**2                    # energi kinetik
        Ep = m * g * s                         # energi potensial
        print(f"{t:<10.1f}{s:<15.2f}{v:<20.2f}{Ek:<20.2f}{Ep:<20.2f}")
        t += dt


# =============================
# KASUS 3
# =============================
print("KASUS 3")
simulasi_gerak(
    m=1.5,
    a=4,
    v0=0,
    h0=20,
    t_total=6,
    dt=1
)
```
