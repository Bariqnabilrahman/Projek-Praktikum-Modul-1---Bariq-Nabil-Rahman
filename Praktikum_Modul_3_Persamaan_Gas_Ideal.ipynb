```python
def konversi_tekanan(nilai, satuan):
    """Mengkonversi tekanan ke satuan SI (Pa)"""
    if satuan == "atm":
        return nilai * 101325
    elif satuan == "Pa":
        return nilai
    else:
        raise ValueError(f"Satuan tekanan '{satuan}' tidak dikenali")

def konversi_volume(nilai, satuan):
    """Mengkonversi volume ke satuan SI (m)"""
    if satuan == "L":
        return nilai / 1000
    elif satuan == "m":
        return nilai
    else:
        raise ValueError(f"Satuan volume '{satuan}' tidak dikenali")

def konversi_suhu(nilai, satuan):
    """Mengkonversi suhu ke satuan SI (K)"""
    if satuan == "C":
        return nilai + 273.15
    elif satuan == "K":
        return nilai
    else:
        raise ValueError(f"Satuan suhu '{satuan}' tidak dikenali")

def main():
    R = 8.314  # Konstanta gas ideal dalam J/(mol·K)

    print("Program Hitung Variabel Persamaan Gas Ideal PV = nRT")
    print("=" * 50)
    print("Pilih variabel yang ingin dihitung:")
    print("1. Tekanan (P)")
    print("2. Volume (V)")
    print("3. Jumlah mol (n)")
    print("4. Suhu (T)")

    try:
        pilihan = input("Masukkan pilihan (1/2/3/4): ")

        if pilihan == "1":  # Hitung Tekanan (P)
            print("\n=== Menghitung Tekanan (P) ===")

            # Input volume
            vol_val = float(input("Masukkan volume: "))
            vol_sat = input("Masukkan satuan volume (L/m): ")
            V = konversi_volume(vol_val, vol_sat)

            # Input jumlah mol
            n = float(input("Masukkan jumlah mol (mol): "))

            # Input suhu
            temp_val = float(input("Masukkan suhu: "))
            temp_sat = input("Masukkan satuan suhu (K/C): ")
            T = konversi_suhu(temp_val, temp_sat)

            # Hitung tekanan
            P = (n * R * T) / V
            print(f"\nHasil perhitungan:")
            print(f"Tekanan (P) = {P:.2f} Pa")

        elif pilihan == "2":  # Hitung Volume (V)
            print("\n=== Menghitung Volume (V) ===")

            # Input tekanan
            press_val = float(input("Masukkan tekanan: "))
            press_sat = input("Masukkan satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(press_val, press_sat)

            # Input jumlah mol
            n = float(input("Masukkan jumlah mol (mol): "))

            # Input suhu
            temp_val = float(input("Masukkan suhu: "))
            temp_sat = input("Masukkan satuan suhu (K/C): ")
            T = konversi_suhu(temp_val, temp_sat)

            # Hitung volume
            V = (n * R * T) / P
            print(f"\nHasil perhitungan:")
            print(f"Volume (V) = {V:.6f} m³")

        elif pilihan == "3":  # Hitung Jumlah Mol (n)
            print("\n=== Menghitung Jumlah Mol (n) ===")

            # Input tekanan
            press_val = float(input("Masukkan tekanan: "))
            press_sat = input("Masukkan satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(press_val, press_sat)

            # Input volume
            vol_val = float(input("Masukkan volume: "))
            vol_sat = input("Masukkan satuan volume (L/m): ")
            V = konversi_volume(vol_val, vol_sat)

            # Input suhu
            temp_val = float(input("Masukkan suhu: "))
            temp_sat = input("Masukkan satuan suhu (K/C): ")
            T = konversi_suhu(temp_val, temp_sat)

            # Hitung jumlah mol
            n = (P * V) / (R * T)
            print(f"\nHasil perhitungan:")
            print(f"Jumlah mol (n) = {n:.5f} mol")

        elif pilihan == "4":  # Hitung Suhu (T)
            print("\n=== Menghitung Suhu (T) ===")

            # Input tekanan
            press_val = float(input("Masukkan tekanan: "))
            press_sat = input("Masukkan satuan tekanan (Pa/atm): ")
            P = konversi_tekanan(press_val, press_sat)

            # Input volume
            vol_val = float(input("Masukkan volume: "))
            vol_sat = input("Masukkan satuan volume (L/m): ")
            V = konversi_volume(vol_val, vol_sat)

            # Input jumlah mol
            n = float(input("Masukkan jumlah mol (mol): "))

            # Hitung suhu
            T = (P * V) / (n * R)
            print(f"\nHasil perhitungan:")
            print(f"Suhu (T) = {T:.2f} K")

        else:
            print("Pilihan tidak valid. Silakan pilih 1, 2, 3, atau 4.")

    except ValueError as e:
        print(f"Error: Input tidak valid - {e}")
    except ZeroDivisionError:
        print("Error: Pembagian dengan nol tidak diperbolehkan.")
    except Exception as e:
        print(f"Error: Terjadi kesalahan - {e}")

if __name__ == "__main__":
    # Jalankan program interaktif
    main()

    print("\n" + "="*50)
```
